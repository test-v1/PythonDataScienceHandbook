```python
import pandas as pd
import numpy as np

# Create a large DataFrame with random data
np.random.seed(0)
data = np.random.randn(100, 30)  # 100 rows and 10 columns
columns = [f'Column_{i}' for i in range(30)]
df = pd.DataFrame(data, columns=columns)

# Function to color negative numbers red and positive numbers green
def color_negative_red(value):
    color = 'red' if value < 0 else 'green'
    return f'color: {color}'

# Apply Styling
styled_df = df.style.applymap(color_negative_red)

# Display the styled DataFrame in a Jupyter Notebook
styled_df
```
